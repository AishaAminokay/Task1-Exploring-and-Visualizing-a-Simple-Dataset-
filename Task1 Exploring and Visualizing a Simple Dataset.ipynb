```python
# Task 1: Exploring and Visualizing a Simple Dataset (Iris Dataset)
# Objective: Load, inspect, analyze, and visualize the Iris dataset

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def main():
    print("===== TASK 1: IRIS DATASET EXPLORATION =====\n")

    # Load Iris dataset from seaborn
    df = sns.load_dataset("iris")

    # Basic Inspection
    print("Dataset Shape:", df.shape)
    print("\nColumn Names:", df.columns.tolist())

    print("\nFirst 5 Rows:\n", df.head())

    # Dataset info
    print("\nDataset Info:\n")
    print(df.info())

    # Summary statistics
    print("\nDataset Description:\n", df.describe())

    # ------------------- VISUALIZATIONS -------------------

    # 1) Scatter Plot (Relationship between features)
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")
    plt.title("Scatter Plot: Sepal Length vs Sepal Width")
    plt.show()

    # 2) Histograms (Distribution)
    df.hist(figsize=(10, 6), edgecolor="black")
    plt.suptitle("Histograms of Iris Features")
    plt.show()

    # 3) Box Plot (Outliers detection)
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df)
    plt.title("Box Plot: Detecting Outliers in Features")
    plt.show()

    print("\n✅ Task 1 Completed Successfully!")

if __name__ == "__main__":
    main()
```
